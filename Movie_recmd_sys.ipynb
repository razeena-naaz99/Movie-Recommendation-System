```python
def genre_list(x):
    gen_list = []
    y= " "
    for i in x:
        gen_list.append(i.get('name'))
    if gen_list == []:
        return np.NaN
    else:
        return (y.join(gen_list))
```

```python
df2['genres'] = df2['genres'].map(lambda x: genre_list(x))
```

```python
df2['genres']
```

```python
i=0
while(i<=3):
    def actor(x):
        casts = []
        for n in x:
            casts.append(n.get('name'))
        if casts == [] or len(casts)<=i:
            return np.NaN
        else:
            return (casts[i])
    df2[f'actor_{i+1}_name'] = df2['cast'].map(lambda x: actor(x))
    i+=1
```

```python
def directors(x):
    d = []
    s = " "
    for i in x:
        if i.get('job') == 'Director':
            d.append(i.get('name'))
    if d == []:
        return np.NaN
    else:
        return (s.join(d))
```

```python
df2['director_name'] = df2['crew'].map(lambda x: directors(x))
df2['director_name']
```

## Building Recommendation System using cosine similarity

```python
da=pd.read_csv('df.csv')
```

```python
da.columns
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# feed combined column to a CountVectorizer() object for getting the count matrix
cv = CountVectorizer()
count_matrix = cv.fit_transform(da["comb"])
```

```python
# obtain the cosine similarity matrix from the count matrix
cosine_sim = cosine_similarity(count_matrix)
```

```python
def get_title_from_index(index):
    return da[da.index == index]["movie_title"].values[0]
def get_index_from_title(title):
    return da[da.movie_title == title].index.values[0]
```

```python
# enter the movie name to get recommendations
#accessing the row corresponding to given movie to find all the similarity scores for that movie and then enumerating over it
movie_user_likes = "the dark knight rises"
movie_index = get_index_from_title(movie_user_likes)
similar_movies = list(enumerate(cosine_sim[movie_index]))
```

```python
# Sort the list similar_movies according to similarity scores in descending order
# Most similar movie to a given movie will be itself, we will discard the first element
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
```

```python
# run a loop to print top 20 similar movies
i=0
print("Top 20 similar movies to "+movie_user_likes+" are:\n")
for element in sorted_similar_movies:
    print(get_title_from_index(element[0]))
    i=i+1
    if i>20:
        break
```
